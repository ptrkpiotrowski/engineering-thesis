```python
# delta plots
for ix, dataset in enumerate(datasets.keys()):
    iter = 0
    figure, axis = plt.subplots(1, 3, figsize=(7.3, 2.6))
    for iy, parameter in enumerate([a1, a2, a3]):
        x = models
        y_str = [result[dataset][parameter] or 0 for result in data.values()]
        y = [Decimal(result[dataset][parameter].replace(",", ".")) or 0 for result in data.values()]

        figure.suptitle(dataset)
        figure.subplots_adjust(top=0.79)

        barlist = axis[iter].bar(x, y, color=color, zorder=2)
        axis[iter].set_title(parameter)
        
        # highlights
        highlights = [list(models).index(obj) for obj in datasets[dataset]] 
        for highlight in highlights:
            barlist[highlight].set_color(second_color)
        if highlights:
            legend_elements = [Patch(facecolor=second_color, label=legend_b)]
            plt.legend(handles=legend_elements, loc=(0.1, 1.16))
        
        axis[iter].grid(linestyle='--', linewidth=0.8, axis='y', zorder=1)
        axis[iter].tick_params('x', labelrotation=90)
        axis[iter].set_ylim(0, 1)
        if iter == 0:
            axis[iter].set_yticklabels([str(tick).replace('.', ',')[:3] for tick in plt.yticks()[0]])
        else:
            axis[iter].set_yticklabels([])
            

        # value labels
        for i in range(len(x)):
            if y[i] < Decimal(0.1):
                text_y = y[i] + Decimal(0.03)
            else:
                text_y = y[i] - Decimal(0.05)
            fontsize = 6.6
            axis[iter].text(i, text_y, y_str[i][:4], ha='center', fontsize=fontsize, color='white' if i in highlights else 'black')

        # plt.title(dataset)
        # plt.ylabel(parameter)
        iter += 1
    figure.savefig(f'delta/{ix}.eps', bbox_inches='tight')
    # plt.show()
    plt.close()
```
